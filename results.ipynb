```python
# Install required libraries
!pip install pandas

# Import necessary libraries
import pandas as pd
from google.colab import files

# Initialize empty DataFrames
df_70_30 = None
df_80_20 = None

# Function to upload and process a CSV file
def upload_and_process():
    global df_70_30, df_80_20
    print("\nðŸ“‚ Please upload a model cross-validation file (70-30 or 80-20)...")
    uploaded = files.upload()
    file_name = list(uploaded.keys())[0]

    # Determine if it's 70-30 or 80-20 based on the file name
    if "70_30" in file_name:
        df_70_30 = pd.read_csv(file_name)
        df_70_30.columns = [col + " (70-30)" for col in df_70_30.columns]
        print("\nâœ… 70-30 Cross-Validation Data Uploaded!")
    elif "80_20" in file_name:
        df_80_20 = pd.read_csv(file_name)
        df_80_20.columns = [col + " (80-20)" for col in df_80_20.columns]
        print("\nâœ… 80-20 Cross-Validation Data Uploaded!")
    else:
        print("\nâš  File name does not indicate 70-30 or 80-20. Please check the file!")

# Call the function to upload both files separately
upload_and_process()  # First upload (either 70-30 or 80-20)
upload_and_process()  # Second upload (remaining file)

# Check if both DataFrames are available before merging
if df_70_30 is not None and df_80_20 is not None:
    # Merge the two DataFrames side by side
    df_combined = pd.concat([df_70_30, df_80_20], axis=1)

    # Save to CSV file
    output_filename = "Model_Cross_Validation_Comparison.csv"
    df_combined.to_csv(output_filename, index=False)

    # Download the CSV file
    files.download(output_filename)

    # Display the final comparison
    print("\nðŸ“Š **Comparison of Model Cross Validation (70-30 vs 80-20) saved as CSV.**")
    display(df_combined)  # Works in Google Colab for better visualization

    print(f"\nâœ… File '{output_filename}' is ready for download!")

else:
    print("\nâš  Please ensure you have uploaded both 70-30 and 80-20 files!")
```
